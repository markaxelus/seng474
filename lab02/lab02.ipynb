```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Generate Seed
np.random.seed(42)

# Generate x values
x = np.linspace(0, 10, 200)

# Generate non-linear y values with and without noise
y_true = np.sin(x) + 0.5*x
y = y_true + np.random.normal(0, 0.2, size=len(x))

# Reshape x for sklearn compatibility
X = x.reshape(-1,1)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.7, label='Data Points')
plt.plot(x, y_true, 'r--', label='True Function')
plt.title('Synthetic Nonlinear Data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Fit standard model
lr_model = LinearRegression()
lr_model.fit(X,y)

# Make predictions
y_pred_lr = lr_model.predict(X)

# Calculate MSE
mse_lr = mean_squared_error(y,y_pred_lr)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.7, label='Data Points')
plt.plot(x, y_true, 'r--', label='True Function')
plt.plot(x, y_pred_lr, 'g-', linewidth=2, label=f'Linear Regression (MSE: {mse_lr:.4f})')
plt.title('Standard Linear Regression on Nonlinear Data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
